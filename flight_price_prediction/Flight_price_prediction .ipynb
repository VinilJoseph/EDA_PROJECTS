# Flight price prediction (EDA & Feature engineering)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_train = pd.read_excel(r"C:\Users\DELL\EDA\flight_price_prediction\flight_price_dataset\Data_Train.xlsx")
df_test = pd.read_excel(r"C:\Users\DELL\EDA\flight_price_prediction\flight_price_dataset\Test_set.xlsx")
```

```python
df = pd.concat([df_train, df_test])
df.head()
```

```python
df.info()
```

```python
# feature engineering process
df['Date'] =  df['Date_of_Journey'].str.split('/').str[0]
df['Month'] =  df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
df.head(2)
```

```python
df['Date_of_Journey'].apply(lambda x:x.split('/')[0])
```

```python
df['Date'] =  df['Date'].astype('int64')
df['Month'] =  df['Month'].astype('int64')
df['Year'] = df['Year'].astype('int64')
```

```python
df.drop('Date_of_Journey', axis = 1, inplace = True)
```

```python
df.head(2)
```

```python
 df['Arrival_Time'] =df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
df.head(2)
```

```python
df.isnull().sum()
```

```python
df['Arrival_Hour'] = df['Arrival_Time'].apply(lambda x:x.split(':')[0]).astype('int64')
df['Arrival_Min'] = df['Arrival_Time'].apply(lambda x:x.split(':')[1]).astype('int64')
```

```python
df.head(2)
```

```python
df.drop('Arrival_Time', axis = 1, inplace = True)
```

```python
df.head(2)
```

```python
df['Dep_Hour'] = df['Dep_Time'].apply(lambda x:x.split(':')[0]).astype('int64')
df['Dep_Min'] = df['Dep_Time'].apply(lambda x:x.split(':')[1]).astype('int64')
```

```python
df.head(2)
```

```python
df.drop('Dep_Time', axis = 1, inplace = True)
```

```python
df.head(2)
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops']  = df['Total_Stops'].map({'non-stop':0, '2 stops':2, '1 stop':1, '3 stops':3, np.NaN:1, '4 stops':4})
```

```python
df.info()
```

```python
df.drop('Route', axis = 1, inplace = True)
```

```python
df.head(2)
```

```python
df['Duration_hr'] = df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
df.drop(6474, axis = 0, inplace = True)
df.drop(2660, axis = 0, inplace = True)
```

```python
df['Duration_hr'] = df['Duration_hr'].astype('int64')
```

```python
df.info()
```

```python
df['Duration_min'] = df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
df['Duration_min'] = df['Duration_min'].astype('int64')
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df['Airline'] = le.fit_transform(df['Airline'])
df['Source'] = le.fit_transform(df['Source'])
df['Destination'] = le.fit_transform(df['Destination'])
df['Additional_Info'] = le.fit_transform(df['Additional_Info'])
```

```python
df['Additional_Info'].unique()
```

```python
df.info()
```

```python
encoded_df = pd.get_dummies(df, columns=['Airline', 'Source', 'Destination', 'Additional_Info'], drop_first=True)
encoded_df.fillna(0, inplace=True)
encoded_df = encoded_df.astype(int)
```
